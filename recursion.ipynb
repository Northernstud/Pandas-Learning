Recursion from 0

```python
#1. Sum of all nat number below n

def sumOfAllNat(n):
    if n == 1:
        return 1

    return n + sumOfAllNat(n-1)

print(sumOfAllNat(3))
```

```python
#2. Fatcotrial 

def fact(n):
    if n==1:
        return 1
    
    return n * fact(n-1)

print (fact(5))
```

```python
#3. Fib 

def fib(n):
    if n == 0: 
        return 0

    if (n == 1 or n == 2):
        return 1
    
    return fib(n-1) + fib(n-2) # so this thing go all the way untill n = 1 and n = 2 and the add up together a lot of times to get that number so they add up from 0s and 1s

print(fib(4))
```

```python
#4. Log base 2

def logb2(n):
    if n == 1: 
        return 0
    else:
        return 1 + logb2(n//2)

print(logb2(16))
```

```python
#5. Decimal to Binary

def decToBin(n):
    if n == 1:
        return "1"
    if n == 0: 
        return "0"
    return str(decToBin(n//2)) + str(n%2)

print(decToBin(22))
```
